```python
display(HTML("""
<style>
/* ===== Colour tokens ===== */
:root{
  --nice-teal:        #228096;
  --nice-teal-75:     #59A0B0;
  --nice-teal-50:     #91C0CB;
  --nice-teal-25:     #C8E0E6;
  --nice-cream:       #DED5CA;
  --nice-cream-75:    #E6E0D7;
  --nice-cream-50:    #EEEAE4;
  --nice-cream-25:    #F7F4F1;
  --nice-blue:        #00436C;
  --nice-blue-75:     #407291;
  --nice-blue-50:     #80A1B5;
  --nice-blue-25:     #BFD0DA;
  --nice-green:       #37906D;
  --nice-tan:         #D07B4D;
  --nice-yellow:      #EAD054;
  --nice-pink:        #EDD8CD;
  --nice-black:       #0e0e0e;
  --nice-white:       #ffffff;
}

/* ===== Base container ===== */
.nice-style{
  font-family: -apple-system, BlinkMacSystemFont, "Segoe UI", Roboto, Arial,
               "Noto Sans", "Liberation Sans", sans-serif;
  line-height: 1.65;
  color: var(--nice-black);
  max-width: 920px;
  margin: 0 auto;
}

/* ===== Report header / banner ===== */
.nice-report-header{
  background: var(--nice-blue);
  color: var(--nice-white);
  padding: 2rem 2.25rem 1.75rem;
  border-radius: 4px;
  margin: 0 0 2rem;
  position: relative;
  overflow: hidden;
}
.nice-report-header::after{
  content: "";
  position: absolute;
  top: 0; right: 0;
  width: 180px; height: 100%;
  background: linear-gradient(135deg, transparent 40%, var(--nice-teal) 100%);
  opacity: 0.35;
}
.nice-report-header h1{
  font-size: 1.65rem;
  font-weight: 700;
  margin: 0 0 0.3rem;
  letter-spacing: -0.01em;
  position: relative;
  z-index: 1;
}
.nice-report-header .subtitle{
  font-size: 0.95rem;
  opacity: 0.88;
  margin: 0;
  font-weight: 400;
  position: relative;
  z-index: 1;
}
.nice-report-header .header-rule{
  display: block;
  width: 64px;
  height: 3px;
  background: var(--nice-teal-50);
  border: none;
  margin: 0.9rem 0 0.7rem;
  position: relative;
  z-index: 1;
}

/* ===== Section headings ===== */
.nice-style h2{
  font-size: 1.45rem;
  font-weight: 700;
  color: var(--nice-blue);
  margin: 2rem 0 0.5rem;
  padding-bottom: 0.35rem;
  border-bottom: 3px solid var(--nice-teal);
}
.nice-style h3{
  font-size: 1.3rem;
  margin: 1.5rem 0 0.4rem;
  font-weight: 700;
  color: var(--nice-blue);
}
.nice-style h4{
  font-size: 1.05rem;
  margin: 1.2rem 0 0.5rem;
  font-weight: 700;
  color: var(--nice-teal);
}

/* ===== Body text & links ===== */
.nice-style p{
  font-size: 1rem;
  margin: 0 0 0.75rem;
}
.nice-style a{
  color: var(--nice-teal);
  text-decoration: underline;
  text-underline-offset: 2px;
}
.nice-style a:hover{
  color: var(--nice-blue);
}

/* ===== Recommendation call-out boxes ===== */
.nice-rec{
  margin: 0.75rem 0 1.25rem;
  padding: 0.15rem 0;
}
.nice-rec-box{
  background: var(--nice-cream-25);
  border-left: 4px solid var(--nice-teal);
  padding: 1rem 1.25rem;
  margin: 0.75rem 0 1.25rem;
  border-radius: 0 4px 4px 0;
}
.nice-rec-box .rec-num{
  display: inline-block;
  font-size: 0.8rem;
  font-weight: 700;
  text-transform: uppercase;
  letter-spacing: 0.04em;
  color: var(--nice-white);
  background: var(--nice-teal);
  padding: 0.15rem 0.6rem;
  border-radius: 3px;
  margin: 0 0 0.6rem;
}
.nice-rec-box p{
  margin: 0;
  font-size: 0.97rem;
}
.nice-rec-box em{
  font-style: italic;
  color: #333;
}
.nice-rec-box ul{
  margin: 0.4rem 0 0.2rem 1.2rem;
  padding: 0;
  font-size: 0.97rem;
  color: #333;
  font-style: italic;
}
.nice-rec-box ul li{
  margin-bottom: 0.2rem;
}

/* ===== Description / narrative text ===== */
.nice-desc{
  margin: 0.5rem 0 1rem;
  padding: 0;
  font-size: 0.97rem;
  color: #444;
}
.nice-desc p{
  margin: 0;
}
             
/* ===== Smaller subsection collapsibles (sex/age/ethnicity/IMD) ===== */
.nice-small{
  font-size: 0.9rem;
}

.nice-small details.nice-collapsible > summary{
  font-size: 1.05rem;
  padding: 0.65rem 1.0rem;
}

.nice-small .collapsible-body{
  padding: 0.75rem 1.0rem 0.35rem;
}

.nice-small .nice-desc,
.nice-small .nice-methods{
  font-size: 0.9rem;
  color: #444;
}
             
/* ===== Figure / image containers ===== */
.nice-figure{
  width: 100%;
  margin: 1rem 0 1.5rem;
  padding: 0;
  background: none;
  border: none;
  border-radius: 0;
  box-sizing: border-box;
}
.nice-figure svg,
.nice-figure img{
  display: block;
  margin: 0 auto;
  width: 100%;
  max-width: 100%;
  height: auto;
}

/* ===== Collapsible / expandable sections ===== */
details.nice-collapsible{
  margin: 1.5rem 0;
  border: none;
  border-radius: 0;
  overflow: visible;
}

details.nice-collapsible > summary{
  display: flex;
  align-items: center;
  gap: 0.5rem;
  cursor: pointer;
  padding: 0.85rem 1.25rem;
  background: var(--nice-cream-25);
  border-bottom: none;
  font-weight: 700;
  font-size: 1.3rem;
  color: var(--nice-blue);
  list-style: none;
  user-select: none;
  transition: background 0.15s;
}

/* hide native marker (WebKit) */
details.nice-collapsible > summary::-webkit-details-marker,
details.nice-sub-collapsible > summary::-webkit-details-marker{
  display: none;
}

/* standard-style triangle (font-independent) */
details.nice-collapsible > summary::before{
  content: "";
  width: 0;
  height: 0;
  border-style: solid;
  border-width: 5px 0 5px 7px;  /* native-like proportions */
  border-color: transparent transparent transparent var(--nice-teal);
  margin-right: 0.45rem;
  transition: transform 0.2s ease;
  flex-shrink: 0;
}

details.nice-collapsible[open] > summary::before{
  transform: rotate(90deg);
}

details.nice-collapsible > summary:hover{
  background: var(--nice-cream-50);
}

/* body padding for top-level collapsible */
details.nice-collapsible > .collapsible-body{
  padding: 1rem 1.25rem 0.5rem;
}

/* --- Nested / sub-collapsible (for breakdowns by sex, age, etc.) --- */
details.nice-sub-collapsible{
  margin: 1rem 0;
  border-left: 3px solid var(--nice-teal-50);
  padding-left: 0;
}

details.nice-sub-collapsible > summary{
  display: flex;
  align-items: center;
  cursor: pointer;
  padding: 0.5rem 0.75rem;
  font-weight: 700;
  font-size: 1.05rem;
  color: var(--nice-teal);
  list-style: none;
  user-select: none;
  transition: color 0.15s;
}

/* nested triangle */
details.nice-sub-collapsible > summary::before{
  content: "";
  width: 0;
  height: 0;
  border-style: solid;
  border-width: 4px 0 4px 6px;
  border-color: transparent transparent transparent var(--nice-teal-50);
  margin-right: 0.35rem;
  transition: transform 0.2s ease;
  flex-shrink: 0;
}

details.nice-sub-collapsible[open] > summary::before{
  transform: rotate(90deg);
}

details.nice-sub-collapsible > summary:hover{
  color: var(--nice-blue);
}

/* body padding for nested collapsible */
details.nice-sub-collapsible > .sub-body{
  padding: 0.5rem 0.75rem 0.25rem;
}
/* ===== Section divider ===== */
.nice-divider{
  border: none;
  border-top: 2px solid var(--nice-cream);
  margin: 2rem 0;
}

/* ===== Introduction paragraph styling ===== */
.nice-intro{
  background: var(--nice-cream-25);
  border-radius: 4px;
  padding: 1.25rem 1.5rem;
  margin: 0 0 1.5rem;
  font-size: 0.97rem;
  line-height: 1.7;
}
.nice-intro p{
  margin: 0 0 0.85rem;
}
.nice-intro p:last-child{
  margin-bottom: 0;
}

/* ===== Methods box ===== */
.nice-methods{
  background: var(--nice-cream-25);
  border-left: 3px solid var(--nice-cream);
  padding: 0.75rem 1rem;
  font-size: 0.93rem;
  color: #555;
  border-radius: 0 4px 4px 0;
}
.nice-methods p{
  margin: 0;
}

</style>
"""))


def collapsible_image(title, path, description=None, max_width=1200, open_by_default=False):
    path = Path(path)
    suffix = path.suffix.lower()
    open_attr = "open" if open_by_default else ""

    if suffix == ".svg":
        content = path.read_text()
    else:
        content = f'<img src="{path}" />'

    desc_html = f'<div class="nice-desc"><p>{description}</p></div>' if description else ""

    display(HTML(f"""
    <details class="nice-collapsible" {open_attr}>
      <summary>{title}</summary>
      <div class="collapsible-body">
        {desc_html}
        <div class="nice-figure" style="max-width:{max_width}px;">
          {content}
        </div>
      </div>
    </details>
    """))


def collapsible_image_html(title, path, description=None, max_width=1200, open_by_default=False):
    svg = Path(path).read_text()

    open_attr = " open" if open_by_default else ""

    desc_html = ""
    if description is not None:
        desc_text = description.strip().replace("\n", " ")
        desc_html = f'<div class="nice-desc"><p>{desc_text}</p></div>'

    return f"""
    <details class="nice-collapsible" {open_attr}>
      <summary>{title}</summary>
      <div class="collapsible-body">
        {desc_html}
        <div class="nice-figure" style="max-width:{max_width}px;">
          {svg}
        </div>
      </div>
    </details>
    """
```
