```python
from pathlib import Path
from IPython.display import HTML, SVG, display
import json
import pandas as pd
import numpy as np
%matplotlib inline

display(HTML("""
<style>
.nice-style{
  font-family: -apple-system,BlinkMacSystemFont,"Segoe UI",Roboto,Arial,"Noto Sans","Liberation Sans",sans-serif;
  line-height: 1.5;
  color: #1a1a1a;
  max-width: 900px;
}
.nice-style h3{
  font-size: 1.35rem;
  margin: 1.2rem 0 0.4rem;
  font-weight: 700;
}
.nice-style h4{
  font-size: 1.05rem;
  margin: 0.2rem 0 0.6rem;
  font-weight: 700;
}
.nice-rec{
  margin: 0.2rem 0 1rem;
  padding: 0.25rem 0 0.25rem;
}
.nice-rec .rec-num{
  display: block;
  font-weight: 700;
  margin: 0 0 0.25rem;
}
.nice-rec p{
  margin: 0;
  font-size: 1rem;
}
             
.nice-small{
  font-size: 0.9rem;
}

.nice-small summary{
  font-size: 1.05rem;
}

.nice-small .nice-desc{
  font-size: 0.9rem;
}

details.guideline-collapsible{
  margin: 1rem 0;
  #border-left: 4px solid #005eb8;
  #padding-left: 0.75rem;
  border-left: none;
  padding-left: 0;
}

details.guideline-collapsible summary{
  padding-bottom: 6px;
  margin-bottom: 6px;
}

details.guideline-collapsible > summary:hover{
  text-decoration: underline;
}

.guideline-description{
  margin: 10px 0;
  color: #444;
}

.guideline-image{
  width: 100%;
  margin-top: 10px;
}

.guideline-image svg,
.guideline-image img{
  display: block;
  margin: 0 auto;
  width: 100%;
  max-width: 100%;
  height: auto;
}

</style>
"""))
             
def collapsible_image(title, path, description=None, max_width=1200, open_by_default=False):
    path = Path(path)
    suffix = path.suffix.lower()
    open_attr = "open" if open_by_default else ""

    if suffix == ".svg":
        content = path.read_text()
    else:
        content = f'<img src="{path}" />'
        
    desc_html = f"<p>{description}</p>" if description else ""

    display(HTML(f"""
    <details class="guideline-collapsible nice-style" {open_attr}>
      <summary>{title}</summary>

      <div class="guideline-description">
        {desc_html}
      </div>

      <div class="guideline-image" style="max-width:{max_width}px;">
        {content}
      </div>
    </details>
    """))

def collapsible_image_html(title, path, description=None, max_width=1200, open_by_default=False):
    svg = Path(path).read_text()
    
    desc_html = ""
    if description is not None:
        desc_html = description.replace("\n", "<br/>")

    return f"""
    <details class="guideline-collapsible nice-style"{' open' if open_by_default else ''}>
      <summary><h4 style="display:inline; margin:0;">{title}</h4></summary>
      <div class="nice-rec">
        <p>{desc_html}</p>
      </div>
      <div class="guideline-image" style="max-width:{max_width}px;">
        {svg}
      </div>
    </details>
    """
```

<div class="nice-style" id="title">
  <h3>Monitoring the incidence and management of gout in England </h3>

  <div class="nice-rec" id="introduction">
    <p>
      Gout is the most common type of inflammatory arthritis worldwide, affecting 1.5 million adults in the UK. People with gout experience attacks of severe joint pain, which can lead to irreversible joint damage and disability if untreated. Gout is also associated with complications such as chronic kidney disease and heart attacks. 
      <br>
      <br>
      There are very effective preventative medications (urate-lowering therapies, such as allopurinol) that can cure the symptoms of gout and prevent complications. However, 
      <a href="https://www.thelancet.com/journals/lanrhe/article/PIIS2665-9913(23)00206-0/fulltext" target="_blank">
        previous studies
      </a>
      found that only a minority of people with gout in the UK were prescribed these treatments.
      <br>
      <br>
      With the goal of improving patient care, the first
      <a href="https://www.nice.org.uk/guidance/ng219" target="_blank">
        NICE guidelines
      </a>
      for the management of gout were published in 2022. NICE guidelines are widely used by primary care clinicians, which is where the majority of gout is managed in the UK. However, we still do not know whether these guidelines are being followed in clinical practice, or if they have improved outcomes for patients with gout.
      <br>
      <br>
      We are using data in the OpenSAFELY platform to show that it is possible to monitor the uptake and impact of NICE guidelines for the management of gout. By demonstrating that this approach is possible for gout, the methodology we develop will be adaptable for monitoring numerous other guidelines and diseases within the NHS.
    </p>
  </div>
</div>
